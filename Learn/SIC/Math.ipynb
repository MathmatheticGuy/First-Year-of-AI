```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
# Câu 1: Vẽ đồ thị hàm Sigmoid
# Tạo một mảng giá trị x từ -10 đến 10 với 400 điểm
x = np.linspace(-10, 10, 400)
# Tính giá trị của hàm sigmoid cho từng giá trị x
sigmoid = 1 / (1 + np.exp(-x))

# Tạo đồ thị
plt.figure(figsize=(8, 5))
plt.plot(x, sigmoid, label=r'$\phi(z) = \frac{1}{1+e^{-z}}$')
plt.title('Sigmoid Function')  # Tiêu đề đồ thị
plt.xlabel('z')  # Nhãn trục x
plt.ylabel(r'$\phi(z)$')  # Nhãn trục y
plt.axhline(0, color='black',linewidth=0.5)  # Vẽ trục hoành (y=0)
plt.axvline(0, color='black',linewidth=0.5)  # Vẽ trục tung (x=0)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)  # Vẽ lưới
plt.legend()  # Hiển thị chú thích
plt.show()  # Hiển thị đồ thị
```

# Câu 2: đạo hàm của $$ f(x) = x^3 - 3x^2 + 2x $$

```python

# Câu 2: Tính đạo hàm của hàm số f(x) = x^3 - 3x^2 + 2x
# Khai báo biến x dùng cho sympy
x = sp.symbols('x')
# Định nghĩa hàm f(x)
f = x**3 - 3*x**2 + 2*x
# Tính đạo hàm của f(x)
f_prime = sp.diff(f, x)
# Chuyển đổi đạo hàm thành hàm số để có thể tính toán giá trị tại các điểm cụ thể
f_prime_func = sp.lambdify(x, f_prime, 'numpy')

# Tạo mảng giá trị x từ -10 đến 10 với 400 điểm
x_vals = np.linspace(-10, 10, 400)
# Tính giá trị của đạo hàm tại các điểm x
f_prime_vals = f_prime_func(x_vals)

# Tạo đồ thị
plt.figure(figsize=(8, 5))
plt.plot(x_vals, f_prime_vals, label=r"$f'(x) = 3x^2 - 6x + 2$")
plt.title('Derivative of f(x) = x^3 - 3x^2 + 2x')  # Tiêu đề đồ thị
plt.xlabel('x')  # Nhãn trục x
plt.ylabel(r"$f'(x)$")  # Nhãn trục y
plt.axhline(0, color='black',linewidth=0.5)  # Vẽ trục hoành (y=0)
plt.axvline(0, color='black',linewidth=0.5)  # Vẽ trục tung (x=0)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)  # Vẽ lưới
plt.legend()  # Hiển thị chú thích
plt.show()  # Hiển thị đồ thị
```

# Câu 3 Đạo hàm của $$ g(x) = sin(e^x) $$

```python
# Câu 3: Tính đạo hàm của hàm hợp g(x) = sin(e^x)
# Định nghĩa hàm g(x)
g = sp.sin(sp.exp(x))
# Tính đạo hàm của g(x)
g_prime = sp.diff(g, x)
# Chuyển đổi đạo hàm thành hàm số để có thể tính toán giá trị tại các điểm cụ thể
g_prime_func = sp.lambdify(x, g_prime, 'numpy')

# Tạo mảng giá trị x từ -2 đến 2 với 400 điểm
x_vals = np.linspace(-2, 2, 400)
# Tính giá trị của đạo hàm tại các điểm x
g_prime_vals = g_prime_func(x_vals)
print(g_prime_vals)
# Tạo đồ thị
plt.figure(figsize=(8, 5))
plt.plot(x_vals, g_prime_vals, label=r"$g'(x) = \cos(e^x) \cdot e^x$")
plt.title(r"Derivative of $g(x) = \sin(e^x)$")  # Tiêu đề đồ thị
plt.xlabel('x')  # Nhãn trục x
plt.ylabel(r"$g'(x)$")  # Nhãn trục y
plt.axhline(0, color='black',linewidth=0.5)  # Vẽ trục hoành (y=0)
plt.axvline(0, color='black',linewidth=0.5)  # Vẽ trục tung (x=0)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)  # Vẽ lưới
plt.legend()  # Hiển thị chú thích
plt.show()  # Hiển thị đồ thị
```
