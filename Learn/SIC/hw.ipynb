## Question 1: Sigmoid graphing

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
# create data points (100 points)
z = np.linspace(-8, 8, 100)
phi_z = sigmoid(z)
```

```python
# plotting
plt.figure(figsize=(5, 3))
plt.plot(z, phi_z, label=r'$\phi(z) = \frac{1}{1 + e^{-z}}$') # plot label using latex
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.ylim(-0.1, 1.1)
plt.xlim(-8, 8)
plt.xlabel('$z$')
plt.ylabel(r'$\phi(z)$')
plt.title('Sigmoid Graph')
plt.legend()
plt.show()
```

## Question 2: Derivative graphing 1

```python
# derivative of function: f'(x)
def df(x):
    return 3 * x**2 - 6 * x + 2
```

```python
# data points (100 points)
x = np.linspace(-10, 10, 100)

# calculate derivative of each x
df_x = df(x)
```

```python
# plotting
plt.figure(figsize=(5, 3))
plt.plot(x, df_x, label=r"$f'(x) = 3x^2 - 6x + 2$")
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.ylim(-100, 100)
plt.xlim(-10, 10)
plt.xlabel('$x$')
plt.ylabel(r"$f'(x)$")
plt.title('$f\'(x) = x^3 - 3x^2 + 2x$')
plt.legend()
plt.show()
```

## Question 3: Derivative graphing 2

```python
# derivative of function: g'(x)
def dg(x):
    return np.cos(np.exp(x)) * np.exp(x)
```

```python
# data points (100 points)
x = np.linspace(-2, 2, 400)

# calculate derivative of each x
dg_x = dg(x)
print(dg_x)
```

```python
# plotting
plt.figure(figsize=(8, 5))
plt.plot(x, dg_x, label=r"$g'(x) = \cos(e^x) \cdot e^x$")
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.ylim(-100, 100)
plt.xlim(-2, 2)
plt.xlabel('$x$')
plt.ylabel(r"$f'(x)$")
plt.title('$g\'(x) = \sin(e^x)$')
plt.legend()
plt.show()
```
